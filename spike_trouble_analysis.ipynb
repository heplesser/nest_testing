```python
import pandas as pd
```

```python
d1 = pd.read_csv('dbvpa_01_all.csv').set_index(['times', 'senders']).sort_index()
d2 = pd.read_csv('dbvpa_02_all.csv').set_index(['times', 'senders']).sort_index()
```

```python
dd = d2 - d1
```

```python
bad_targets = list(dd.loc[dd.I_syn_ex != 0].index.get_level_values('senders'))
```

```python
sources = [10, 49, 41, 41, 9, 44, 19, 46, 9, 25, 11, 33, 46, 37,
               36, 27, 45, 29, 15, 27, 21, 50, 27, 38, 3, 5, 38, 3,
               41, 49, 42, 37, 36, 45, 5, 3, 21, 29, 9, 30, 34, 40,
               35, 44, 41, 48, 44, 27, 36, 47]
```

```python
tgt_2_src = {i+1: s for i, s in enumerate(sources)}
```

```python
bad_sources = {tgt_2_src[t] for t in bad_targets}
```

```python
bad_sources
```

```python
good_sources = set(range(1, 51))-bad_sources
```

```python
bad_conns = [(s, i+1) for i, s in enumerate(sources) if i+1 in bad_targets]
good_conns = [(s, i+1) for i, s in enumerate(sources) if i+1 not in bad_targets]
```

```python
bad = pd.DataFrame(bad_conns, columns=['Src', 'Tgt'])
bad['OK'] = False
good = pd.DataFrame(good_conns, columns=['Src', 'Tgt'])
good['OK'] = True
conns = pd.concat([bad, good], ignore_index=True)
```

```python
conns['Sv'] = conns.Src % 4
conns['Tv'] = conns.Tgt % 4

conns['Sr'] = conns.Sv % 2
conns['Tr'] = conns.Tv % 2

conns['St'] = (conns.Sv // 2) % 2
conns['Tt'] = (conns.Tv // 2) % 2
```

```python
conns
```

```python
conns.loc[~conns.OK, 'Src':'Tgt'].values.to_list()
```
